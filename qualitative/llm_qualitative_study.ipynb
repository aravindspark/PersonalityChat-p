```python
from openai import OpenAI
from openai import APIConnectionError
import os

api_models_list = [
    "llama-3-8b-instruct",
    "llama-3.3-70b-instruct",
    "qwen-2.5-7b-instruct", 
    "qwen-2.5-72b-instruct",
]

api_base_url = ""

class OpenAIClient:
    def __init__(self, api_key, model_name=None):
        if model_name is None or model_name not in api_models_list:
            return  exit(f"Please provide a valid model name from the list: {api_models_list}")

        self.client = OpenAI(base_url=api_base_url, api_key=api_key)
        self.model_name = model_name
     
    def get_response(self, statement):
        
        system_prompt = f"""You are an helpful assitant who can only reply to the user questions Honestly and Truthfully. 
        Your replies should be detailed and elaborate."""

        try:
            response = self.client.chat.completions.create(
                        model = self.model_name,
                        messages=[
                            {"role": "system", "content": system_prompt},
                            {"role": "user", "content": statement},
                        ],
                        max_tokens=2048,
                        temperature=0.1,
                        top_p=0.1,
                        )
        except APIConnectionError as e:
            print(f"API Connection Error: {e}")
            # Retry in 2 seconds
            # time.sleep(2)
            return None
        
        except Exception as e:
            print(f"API Error: failed to get response from {self.model_name}: {e}")
            return None
        
        return response.choices[0].message.content
```

```python
responses = []

def run_qualitative_study(client, question):
    response = client.get_response(question)
    responses.append({ 
                      "model": client.model_name, 
                      "question": question,
                      "response": response
                      })
```

```python
llm_questions = [
"When you receive a prompt that allows for many possible directions, how do you decide whether to respond with something imaginative, structured, cautious, or exploratory?"

,"How do you balance being precise and orderly in your responses with allowing flexibility or creativity when the situation seems open-ended?"

,"In what ways do you adjust the emotional tone of your replies depending on how the prompt is framed or the needs of the user?"

,"When a prompt asks for collaboration or negotiation of ideas, how do you navigate between asserting a perspective and accommodating the other’s expectations?"

,"How do you typically handle uncertainty or ambiguity in a request—do you clarify, generate multiple possibilities, or take a more cautious approach?"
]
```

```python

# model_list = ["llama-3.3-70b-instruct", "qwen-2.5-72b-instruct","qwen-2.5-7b-instruct"]
model_list = ["qwen-2.5-72b-instruct"]


for model_name in model_list:
    client = OpenAIClient(api_key=os.getenv("LLM_API_KEY"), model_name=model_name)
    for question in llm_questions:
        run_qualitative_study(client, question)
```

#### record it  to a file

```python
with open("qualitative_reponses_llm.json", "w") as f:
    import json
    json.dump(responses, f, indent=4)
```
