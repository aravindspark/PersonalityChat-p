```python
import json
import pandas as pd
import re

reverse_score = { 1: 5, 2: 4, 3: 3, 4: 2, 5: 1 }

choices_to_score = { 'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5 }

text_to_score = {
    'disagree strongly': 1,
    'disagree a little': 2,
    'neither agree nor disagree': 3,
    'agree a little': 4,
    'agree strongly': 5
}
```

```python
bfi_data = pd.read_csv("questionaries_BFI.csv")
bfi_data["number"] = bfi_data["number"].astype(int)
```

```python
bfi_data = bfi_data.sample(frac=1).reset_index(drop=True)
print(bfi_data.head())
```

```python
grouped = bfi_data.groupby("category")
for category, group in grouped:
    print(f"Category: {category}")
    for index, row in group.iterrows():
        question = row['question']
        number = row['number']
        print(f"  Question {number}: {question}")
```

```python
def process_response(response, resp_fmt="number"):
    for answer in response.splitlines():
        answer = answer.strip()
        print(f"Processing answer: {answer}")
        if not answer:
            continue
        chunks = re.match(r'^(\d{1,2})[.:]?\s*(.*)$', answer)
        if chunks is None:
            continue
        
        q_num = chunks.group(1)
        resp_txt = chunks.group(2).strip()
        
        try:
            if resp_txt == "":
                resp_val = 0
            else:
                if resp_fmt == "number":
                    resp_val = int(resp_txt)
                elif resp_fmt == "label":
                    resp_val = choices_to_score.get(resp_txt.lower(), 0)
                elif resp_fmt == "text":
                    resp_val = text_to_score.get(resp_txt.lower(), 0)
                    
            print(f"Question Number: {q_num}, Response Value: {resp_val}")
        except ValueError:
            continue
```

```python
response = "1. Agree Strongly\n 3. disagree a little"  # Example response for testing
process_response(response, resp_fmt="text")
print()
response = "1. 5\n2. 8\n 5. 1"  # Example response for testing
process_response(response, resp_fmt="number")
```
